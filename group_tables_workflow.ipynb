```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
file_path = "./flt_tables.xlsx"

df = pd.read_excel(file_path, sheet_name="Tables", engine="openpyxl")
```

```python
df.head()
```

```python
grouped_df = df[df["Required for current reporting"] == "Yes"].groupby(['Table Type', 'Frequency', 'Required for current reporting']).agg({'Table Name': lambda x: ', '.join(x)}).reset_index()
grouped_df
```

```python
grouped_df = df.groupby(['Table Type', 'Frequency', 'Required for current reporting']).agg({'Table Name': lambda x: ', '.join(x)}).reset_index()
grouped_df
```

```python
# Define the path where you want to save the Excel file
file_path = './grouped_tables.xlsx'

# Save the DataFrame to an Excel file
grouped_df.to_excel(file_path, index=False, engine='openpyxl')
```
